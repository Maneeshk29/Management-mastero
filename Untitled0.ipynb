```python
#EmployeeLevel
const readlineSync = require('readline-sync');

// Function to rank employees
function rankEmployees(employees) {
    return employees.map(employee => {
        let level = 'Entry-level';
        if (employee.experience >= 8) level = 'Executive';
        else if (employee.experience >= 3) level = 'Mid-level';
        return { ...employee, level };
    });
}

// Ask for the number of employees
const numEmployees = parseInt(readlineSync.question("Enter the number of employees: "));

// Input data for each employee
const formattedEmployeesData = Array.from({ length: numEmployees }, () => ({
    name: readlineSync.question("Enter employee's full name: "),
    experience: parseInt(readlineSync.question("Enter employee's years of experience: "))
}));

// Rank employees
const rankedEmployeesData = rankEmployees(formattedEmployeesData);

// Output the ranked employees data
rankedEmployeesData.forEach(employee => {
    console.log(`${employee.name} - Experience: ${employee.experience} years, Level: ${employee.level}`);
});
```

```python
#YearsatCompany
const readlineSync = require('readline-sync');

// Function to rank employees
function rankEmployees(employees) {
    return employees.map(employee => ({
        full_name: employee.full_name,
        years_at_company: employee.years_at_company,
        points: employee.years_at_company >= 10 ? 5 : employee.years_at_company >= 5 ? 3 : employee.years_at_company < 2 ? 1 : 0
    }));
}

// Prompt the user for input
const numEmployees = parseInt(readlineSync.question("Enter the number of employees: "));
const employeesData = Array.from({ length: numEmployees }, (_, i) => ({
    full_name: readlineSync.question(`Enter the full name of employee ${i + 1}: `),
    years_at_company: parseInt(readlineSync.question(`Enter the number of years at company for employee ${i + 1}: `))
}));

// Call the rankEmployees function and print the ranked employees
rankEmployees(employeesData).forEach(employee => console.log(`${employee.full_name} - Points: ${employee.points}`));
```

```python
#Possession of Advanced Specializations and Graduate Degrees
const readlineSync = require('readline-sync');

// Function to assign score based on qualifications
function assignScore(qualifications) {
    const keys = ["PhD", "Master's", "specialization", "Undergraduate", "HS Diploma"];
    return keys.reduce((acc, key) => acc + (qualifications[key] === "yes" ? keys.indexOf(key) + 1 : 0), 0);
}

// Function to gather qualifications for a single employee
function gatherQualifications(fullName) {
    const qualifications = {};
    ["PhD", "Master's", "specialization", "Undergraduate", "HS Diploma"].forEach(key => {
        qualifications[key] = readlineSync.question(`Do you have a ${key}, ${fullName}? (yes/no): `).toLowerCase();
    });
    return qualifications;
}

// Main function to process multiple employees
function processEmployees() {
    const numEmployees = parseInt(readlineSync.question("Enter the number of employees: "));
    for (let i = 0; i < numEmployees; i++) {
        const fullName = readlineSync.question(`Enter full name of employee ${i + 1}: `);
        const qualifications = gatherQualifications(fullName);
        const score = assignScore(qualifications);
        console.log(`${fullName}'s qualification score is: ${score}`);
    }
}

// Entry point
processEmployees();
```

```python
#UTO/PTO
const readlineSync = require('readline-sync');

class Employee {
    constructor(name, tier, ptoBalance = null, utoBalance = null) {
        this.name = name;
        this.tier = tier;
        this.ptoBalance = ptoBalance !== null ? ptoBalance : this.getPtoAllowance();
        this.utoBalance = utoBalance !== null ? utoBalance : this.getUtoAllowance();
        this.ptoAllowance = this.getPtoAllowance();
        this.utoAllowance = this.getUtoAllowance();
    }

    getPtoAllowance() {
        switch (this.tier) {
            case "executive":
                return 30; // Example: Executive employees get 30 days of PTO per year
            case "mid-level":
                return 20; // Example: Mid-level employees get 20 days of PTO per year
            case "entry-level":
                return 15; // Example: Entry-level employees get 15 days of PTO per year
            default:
                return 0; // Default allowance for unrecognized tiers
        }
    }

    getUtoAllowance() {
        switch (this.tier) {
            case "executive":
                return 10; // Example: Executive employees get 10 days of UTO per year
            case "mid-level":
                return 5;  // Example: Mid-level employees get 5 days of UTO per year
            case "entry-level":
                return 0;  // Example: Entry-level employees do not get UTO
            default:
                return 0; // Default allowance for unrecognized tiers
        }
    }

    requestPto(days) {
        if (this.ptoBalance >= days) {
            this.ptoBalance -= days;
            console.log(`${this.name} has taken ${days} days of PTO. Remaining PTO balance: ${this.ptoBalance}`);
        } else {
            console.log(`Insufficient PTO balance for ${this.name} to take ${days} days of PTO.`);
        }
    }

    requestUto(days) {
        if (this.utoBalance >= days) {
            this.utoBalance -= days;
            console.log(`${this.name} has taken ${days} days of UTO. Remaining UTO balance: ${this.utoBalance}`);
        } else {
            console.log(`Insufficient UTO balance for ${this.name} to take ${days} days of UTO.`);
        }
    }
}

// Get the number of employees from the user
const numEmployees = parseInt(readlineSync.question("How many employees does your company have?"));
const employees = [];

// Get employee details from the user
for (let i = 0; i < numEmployees; i++) {
    const fullName = readlineSync.question(`Enter the full name of employee ${i + 1}: `);
    const tier = readlineSync.question(`Enter the tier of employee ${i + 1} (executive/mid-level/entry-level):`);

    // Create an instance of Employee and add it to the array
    const employee = new Employee(fullName, tier);
    employees.push(employee);
}

// Example usage: Request PTO and UTO for each employee
for (const employee of employees) {
    employee.requestPto(5);
    employee.requestUto(8);
}
```

```python
#BackgroundCheck
const readlineSync = require('readline-sync');

function pointSystem() {
    const pointsDict = {};

    const numPeople = parseInt(readlineSync.question("Enter the number of people: "));

    for (let i = 0; i < numPeople; i++) {
        const name = readlineSync.question(`Enter the full name of person ${i + 1}: `);
        const backgroundCheck = readlineSync.question(`Did ${name} pass the background check? (yes/no): `).toLowerCase();

        if (backgroundCheck === 'yes') {
            pointsDict[name] = 5;
        } else if (backgroundCheck === 'no') {
            pointsDict[name] = -5;
        } else {
            console.log("Invalid input. Please enter 'yes' or 'no'.");
            return;
        }
    }

    console.log("\nPoints:");
    for (const [name, points] of Object.entries(pointsDict)) {
        console.log(`${name}: ${points}`);
    }
}

// Call the function to start the point system
pointSystem();

```

```python
#ManagerReview
const readlineSync = require('readline-sync');

function getRatings(fullName) {
    console.log(`\nPlease rate ${fullName}'s performance from 1 to 5 for each category:`);
    const ratings = {};
    const categories = [
        "Quality of Work",
        "Meeting Project Deadlines",
        "Leadership potential",
        "Teamwork and collaboration",
        "Professionalism and ethic"
    ];
    categories.forEach(category => {
        while (true) {
            const rating = parseInt(readlineSync.question(`${category}: `));
            if (rating < 1 || rating > 5 || isNaN(rating)) {
                console.log("Please enter a rating between 1 and 5.");
            } else {
                ratings[category] = rating;
                break;
            }
        }
    });
    return ratings;
}

function calculateAverageScore(ratings) {
    const totalScore = Object.values(ratings).reduce((acc, curr) => acc + curr, 0);
    const numCategories = Object.keys(ratings).length;
    const averageScore = totalScore / numCategories;
    return Math.round(averageScore);
}

function main() {
    const numPeople = parseInt(readlineSync.question("Enter the number of people to review: "));
    const peopleScores = {};

    for (let i = 0; i < numPeople; i++) {
        const fullName = readlineSync.question(`\nEnter the full name of person ${i + 1}: `);
        const ratings = getRatings(fullName);
        const averageScore = calculateAverageScore(ratings);
        peopleScores[fullName] = averageScore;
    }

    console.log("\nAverage Scores:");
    Object.entries(peopleScores).forEach(([fullName, score]) => {
        console.log(`${fullName}: ${score}`);
    });
}

main();
```
